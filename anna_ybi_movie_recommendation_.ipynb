**Get Feature Text Conversion to Tokens**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer()
```

```python
x=tfidf.fit_transform(x)
```

```python
x.shape
```

```python
print(x)
```

#Get Similarity Score using Cosine Similarity

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
Similarity_Score=cosine_similarity(x)
```

```python
Similarity_Score
```

```python
Similarity_Score.shape
```

#Get Movie Name as Input from User and Validate for Closest Spelling

```python
fav_movie_name=input('Enter your favourite movie name:')
```

```python
all_mv_list=mv['Movie_Title'].tolist()
```

```python
import difflib
```

```python
mv_recommned=difflib.get_close_matches(fav_movie_name,all_mv_list)
```

```python
print(mv_recommned)
```

```python
close_match=mv_recommned[0]
```

```python
print(close_match)
```

```python
index_of_close_match=mv[mv.Movie_Title==close_match]['Movie_ID'].values[0]
```

```python
print(index_of_close_match)
```

```python
#getting list of similler movies
recommndation_score=list(enumerate(Similarity_Score[index_of_close_match]))
```

```python
print(recommndation_score)
```

```python
len(recommndation_score)
```

#Get All Movies Sort Based on Recommendation Score wrt Favourite Movie

```python
#sorting movies based on similarity score
sorted_smiliar_mv=sorted(recommndation_score,key=lambda x:x[1],reverse=True)
```

```python
print(sorted_smiliar_mv)
```

```python
#print top 30 simillar movies based on index
print('Top 30 movies suggested for you')
i=1
for movie in sorted_smiliar_mv:
  index=movie[0]
  title_from_index=mv[mv.index==index]['Movie_Title'].values[0]
  if(i<31):
    print(i,'.',title_from_index)
    i+=1
```

#top 10 movies recommendation

```python
movie_name=input('Enter your favourite movie name:')
print(movie_name)
list_of_all_mv=mv['Movie_Title'].tolist()
find_close_match=difflib.get_close_matches(movie_name,list_of_all_mv)
close_match=find_close_match[0]
index_of_mv=mv[mv.Movie_Title==close_match]['Movie_ID'].values[0]
recommondation_score=list(enumerate(Similarity_Score[index_of_mv]))
sorted_recommondation_score=sorted(recommondation_score,key=lambda x:x[1],reverse=True)
print('top 10 movies suggested for you:\n')
i=1
for movie in sorted_recommondation_score:
  index=movie[0]
  title_from_index=mv[mv.Movie_ID==index]['Movie_Title'].values
  if(i<11):
    print(i,'.',title_from_index)
    i+=1

```
